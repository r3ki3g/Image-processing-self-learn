```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


%matplotlib nbagg
```

```python
image_location = "D:/ENTC/SEM_4/EN2550 - Fundamentals of Image Processing and Machine Vision/~images/spider.png"
img = cv.imread(image_location)
img = cv.cvtColor(img,cv.COLOR_RGB2BGR)
```

```python
fig,ax = plt.subplots(1)
ax.imshow(img)
plt.show()
```

```python
img_part = img[48:220+1,377:512+1,:]

# y:48,220
# x : 377,512


fig,ax = plt.subplots(1)
ax.imshow(img_part)
plt.show()
```

```python
img_part[40,41]
```

```python
red = np.zeros(img_part.shape,dtype=np.uint8)
green = np.zeros(img_part.shape,dtype=np.uint8)
blue = np.zeros(img_part.shape,dtype=np.uint8)

red[:,:,0] = img_part[:,:,0]
green[:,:,1] = img_part[:,:,1]
blue[:,:,2] = img_part[:,:,2]



fig,ax = plt.subplots(3)
ax[0].imshow(red)
ax[1].imshow(green)
ax[2].imshow(blue)
plt.show()
```

```python
all_together = red + green + blue
fig,ax = plt.subplots(1)
ax.imshow(all_together)
plt.show()
```
