```python
def f(x):
    return x**2 - 6*x + 9


def g(x):
    return 2*x - 6


alpha = 0.1

x = 1
x_hist =np.array(x)
fx_hist = np.array(f(x))
for i in range(20):
    x = x - alpha * g(x)
    x_hist = np.append(x_hist,x)
    fx_hist = np.append(fx_hist,f(x))
    
    
    
    
#plotting

t = np.linspace(-4,4,100)
y = np.array([f(x) for x in t])

fig,ax = plt.subplots(1,figsize=(20,10))
ax.plot(t,y)
ax.scatter(x_hist,fx_hist,marker="o")


plt.show()
```

## Trying gradient descent

```python
dataPoints = []
for i in range(1000):
    w_true = np.array([1,2,3,4,5])
    x_instance = np.random.rand(5)
    y_instance = w_true.dot(x_instance) + np.random.rand(1)/10
    dataPoints.append((x_instance,y_instance))


    
d = 5 # w1 x1 + w2 x2 + w3 x3 + w4 x4  + w5 x5 = y_pred

def F(w):
    loss = sum((w.dot(x)-y)**2 for x,y in dataPoints)/len(dataPoints)
    return loss

def dF(w):
    gradient = np.sum(2*(w.dot(x) - y)*x for x,y in dataPoints)/len(dataPoints)
    return gradient



#optimization

lr = 0.1 #learning rate AKA step size
w = np.zeros(d)
for i in range(400):
    w -= lr*dF(w)
    print("iterration {} : w = {}".format(i,str(w)))
    





```

## Trying stochastic gradient discent

Randomly selecting just one data sample per each iteration. (Rather than all the data points per iteration)

If the data et has too much noise this method will not work well!

```python
dataPoints = []
for i in range(1000):
    w_true = np.array([1,2,3,4,5])
    x_instance = np.random.rand(5)
    y_instance = w_true.dot(x_instance) + np.random.rand(1)/10 
    dataPoints.append((x_instance,y_instance))


    
d = 5 # w1 x1 + w2 x2 + w3 x3 + w4 x4  + w5 x5 = y_pred

def F(w):
    x,y = dataPoints[random.randint(0, len(dataPoints)-1)]
    loss = (w.dot(x)-y)**2
    return loss

def dF(w):
    x,y = dataPoints[random.randint(0, len(dataPoints)-1)] #selecting a randome data point
    gradient = 2*(w.dot(x) - y)*x 
    return gradient



#optimization

lr = 0.1 #learning rate AKA step size
w = np.zeros(d)
for i in range(400):
    w -= lr*dF(w)
    print("iterration {} : w = {}".format(i,str(w)))
    





```

## Designing a function predictor

Let's find the polynomial expansion for sin(x) with gradient descent

```python
dataPoints = []
for i in range(100):
    
    x = 2*(0.5-np.random.rand(1)) * (2*np.pi) * 2 # the x values are between - 4 pi and + 4 pi
    y_instance = np.sin(x)
    x_into_powers = [1,x,x**2/(1*2),x**3/(1*2*3),x**4/(1*2*3*4),x**5/(1*2*3*4*5),x**6/(1*2*3*4*5*6),x**7/(1*2*3*4*5*6*7)]
    dataPoints.append((x_into_powers,y_instance))




    
d = 8 

def F(w):
    loss = sum((w.dot(x)-y)**2 for x,y in dataPoints)/len(dataPoints)
    return loss

def dF(w):
    gradient = np.sum(2*(w.dot(x) - y)*x for x,y in dataPoints)/len(dataPoints)
    return gradient.astype(np.float32)



#optimization

lr = 0.05 #learning rate AKA step size
w = np.ones(d)
for i in range(400):
    #lrVar = lr / (1+np.sqrt(np.sum((w-np.ones(d))**2))) # same lr for each parameter
    lrVar = lr / (1+ (w-np.ones(d))**2) # different lr for each parameter
    w -= lr*dF(w)
    print("iterration {} : w = {} \n lrVar = {}".format(i+1,str(w),lrVar))
    




    
```
