## Histogram equalization

```python
img = cv.imread("D:\ENTC\SEM_4\EN2550 - Fundamentals of Image Processing and Machine Vision\~images\shells.tif",0)
assert img is not None;

hist =  cv.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()


normalizedHist = hist/sum(hist)
plt.plot(normalizedHist)
plt.show()

#equialization
t=np.zeros(len(normalizedHist))
for i in range(len(normalizedHist)):
    if not i:
        t[i] = normalizedHist[i] *255
    else:
        t[i] = t[i-1] + normalizedHist[i] *255
        
plt.plot(t)
plt.show()

imgHistEquialized = cv.LUT(img,t)
plt.imshow(imgHistEquialized,cmap="gray")
plt.show()
```

## Gaussian blur and box kernel blur


```python
img = cv.imread("D:\ENTC\SEM_4\EN2550 - Fundamentals of Image Processing and Machine Vision\~images\cat.jpg")
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)


ksize=15
avgKernel = 1./ksize**2 * np.ones((ksize,ksize))
imgBoxAvg = cv.filter2D(img,-1,avgKernel)
imgGaussianBlur = cv.GaussianBlur(img,(ksize,ksize),7)


fig,ax = plt.subplots(1,3,figsize=(28,4))
ax[0].imshow(img)
ax[1].imshow(imgBoxAvg)
ax[2].imshow(imgGaussianBlur)
```

```python
img =  cv.imread("D:/ENTC/SEM_4/EN2550 - Fundamentals of Image Processing and Machine Vision/~images/spider.png",0)
assert img is not None
sobel_vertical = np.array([[-1,-2,-1],[0,0,0],[1,2,1]],dtype=np.float32)

horizontalEdges = cv.filter2D(img,-1,sobel_vertical)

plt.imshow(np.abs(horizontalEdges),cmap="gray")
plt.show()
```

```python
sobel_horizontal = sobel_vertical.T

verticalEdges = cv.filter2D(img,-1,sobel_horizontal)

plt.imshow(np.abs(verticalEdges),cmap="gray")
plt.show()
```

```python
edges = (verticalEdges**2+horizontalEdges**2)**0.5
plt.imshow(edges,cmap="gray")
plt.show()
```

```python
#testing for seperabilty of gaussian kernel


g1D = cv.getGaussianKernel(101,20)

g2D = g1D @ g1D.T

plt.imshow(g2D,cmap="gray")
plt.show()



#differentiating
prewit_h = np.array([[-1.0,1],
                    [-1.0,1],
                    [-1.0,1]],dtype=np.float32)
d_g2D  = cv.filter2D(g2D,-1,prewit_h)


plt.imshow(d_g2D,cmap="gray")
plt.show()
```

## Harris corner detection


```python
img = cv.imread("D:/ENTC/SEM_4/EN2550 - Fundamentals of Image Processing and Machine Vision/~images/sudoku.png",1)
assert img is not None

I = np.float32(cv.cvtColor(img,cv.COLOR_BGR2GRAY))

sobel_v = np.array([[-1,-2,-1],[0,0,0],[1,2,1]]).astype(np.float32)
sobel_h = sobel_v.T

Ix = cv.filter2D(I,-1,sobel_v)
Iy = cv.filter2D(I,-1,sobel_h)

ksize = 11

m11 = cv.GaussianBlur(Ix**2,(ksize,ksize),3)
m12 = cv.GaussianBlur(Ix*Iy,(ksize,ksize),3)
m21 = m12
m22 = cv.GaussianBlur(Iy**2,(ksize,ksize),3)

alpha = 0.04
r = (m11*m22-m12*m21) - alpha*(m11+m22)**2
output = r>1e7


plt.imshow(img,cmap="gray")
plt.show()

plt.imshow(output,cmap="gray")
```
